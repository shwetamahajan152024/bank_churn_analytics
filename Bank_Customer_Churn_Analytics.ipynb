```python
#Phase 4 -SQL queries
import pandas as pd
from sqlalchemy import create_engine

#create engine
engine= create_engine('sqlite:///bank_churn.db')
```

```python
#Query 1: Overall churn rate
query1= "SELECT AVG(Exited)*100 AS churn_rate From churn_data "
result1=print(pd.read_sql(query1, con=engine))
print("Overall Churn Rate:")
print(result1)
```

```python
query2="""SELECT Geography, COUNT(*) AS Total_Customers, SUM(Exited) AS Churned, ROUND(AVG(Exited)*100,2) AS Churn_Rate
From churn_data
GROUP BY Geography
ORDER BY Churn_Rate DESC"""
result2=pd.read_sql(query2, con=engine)
print("Churn by Geography:")
print(result2)
```

```python
#Query 3: Churn by Age group
query3= """sELECT
            CASE
                WHEN Age < 30 THEN 'Under 30'
                WHEN Age BETWEEN 30 AND 40 THEN '30-40'
                WHEN Age BETWEEN 40 AND 50 THEN '40-50'
                WHEN Age BETWEEN 50 AND 60 THEN '50-60'
                ELSE 'Over 60'
            END AS Age_Group,
            COUNT(*) AS Total_Customers,
            SUM(Exited) AS Churned,
            ROUND(AVG(Exited)*100,2) AS Churn_Rate
        FROM churn_data
        GROUP BY Age_Group
        ORDER BY Churn_Rate DESC"""
result3=pd.read_sql(query3, con=engine)
print("Churn by Age Group:")
print(result3)
```

```python
#Query 4: Churn by Balance Group
query4= """SELECT
            CASE
                WHEN Balance = 0 THEN 'Zero Balance'
                WHEN Balance <50000 THEN 'Low (<50k)'
                WHEN Balance BETWEEN 50000 AND 100000 THEN 'Medium (50k-100k)'
                ELSE 'High (>100k)'
            END AS Balance_Group,
            COUNT(*) AS Total_Customers,
            SUM(Exited) AS Churned,
            ROUND(AVG(Exited)*100,2) AS Churn_Rate
        FROM churn_data
        GROUP BY Balance_Group
        ORDER BY Churn_Rate DESC"""
result4=pd.read_sql(query4, con=engine)
print("Churn by Balance Group:")
print(result4)
```

```python
#Query 5: Active vs Inactive Memeber Churn
query5= """SELECT
            CASE
                WHEN IsActiveMember =1 THEN 'Active'
                ELSE 'Inactive' END AS Member_Status,
                COUNT(*) AS Total_Customers,
                SUM(Exited) AS Churned,
                ROUND(AVG(Exited)*100,2) AS Churn_Rate
            FROM churn_data
            GROUP BY IsActiveMember
            ORDER BY Churn_Rate DESC"""
result5=pd.read_sql(query5, con=engine)
print("Churn by Member Status:")
print(result5)
```

```python
#Query 6: High Risk Customer Segment
query6="""SELECT Geography, Age, Balance,
           ROUND(AVG(Exited)*100,2)AS Churn_Rate,
           Count(*) AS Total_Customers
        FROM churn_data
        WHERE IsActiveMember=0
        GROUP BY Geography,
        CASE
            WHEN Age<30 THEN 'Under 30'
            WHEN Age BETWEEN 30 AND 45 THEN '30-45'
            WHEN Age BETWEEN 46 AND 60 THEN '46-60'
            ELSE 'Above 60'
        END
        HAVING COUNT(*)>50
        ORDER BY Churn_Rate DESC
        LIMIT 10"""
result6=pd.read_sql(query6, con=engine)
print("High Risk Segment:")
print(result6)
```
